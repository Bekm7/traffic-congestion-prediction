```python
from datetime import datetime
from meteostat import Hourly, Stations
import pandas as pd

# Set time period
start = datetime(2025, 1, 1)
end = datetime(2025, 1, 31, 23, 59)

# Find weather station for NYC (e.g., Central Park)
stations = Stations()
station = stations.nearby(40.7812, -73.9665).fetch(1)

# Fetch hourly data
data = Hourly(station.index[0], start, end)
df_weather = data.fetch()

# Save to CSV
df_weather.to_csv("nyc_jan2025_hourly_weather.csv")

# Preview
print(df_weather.head())
```

```python
import numpy as np

# Parse date properly
snow_df = pd.read_csv("D:/traffic-congestion-predictor/data/raw/snowfall_data.csv")
snow_df['date'] = pd.to_datetime("2025-" + snow_df['Date (2025)'], format='%Y-%m-%d', errors='coerce')

# Step 1: Encode snow as 1 or 0
def encode_snow(row):
    snow = row['Snow (mm)']
    ground = row['Snow on Ground (mm)']
    if isinstance(ground, str) and 'trace' in ground.lower():
        return 1
    if pd.to_numeric(snow, errors='coerce') > 0:
        return 1
    if pd.to_numeric(ground, errors='coerce') > 0:
        return 1
    return 0

snow_df['snowed'] = snow_df.apply(encode_snow, axis=1)

# Step 2: Expand each date into 24 hourly timestamps
expanded_rows = []

for _, row in snow_df.iterrows():
    if pd.isnull(row['date']):
        continue
    for hour in range(24):
        dt = pd.Timestamp(row['date']) + pd.Timedelta(hours=hour)
        expanded_rows.append({'hour': dt, 'snowed': row['snowed']})

hourly_snow_df = pd.DataFrame(expanded_rows)

# Step 3: Now merge with weather_df on 'hour'
weather_df['hour'] = pd.to_datetime(weather_df['hour'])

merged_df = pd.merge(weather_df, hourly_snow_df, on='hour', how='left')
merged_df['snowed'] = merged_df['snowed'].fillna(0).astype(int)

# Step 4: Save the final merged data
merged_df.to_csv("D:/traffic-congestion-predictor/data/processed/weather_with_snow.csv", index=False)
```
