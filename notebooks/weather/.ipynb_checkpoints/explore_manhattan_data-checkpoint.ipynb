```python
import pandas as pd

df = pd.read_parquet('D:/traffic-congestion-predictor/data/raw/yellow_tripdata_2025-01.parquet')
df.sample(5)
```

```python
df = df.sample(n=50000, random_state=42)
```

```python
df.info()
```

```python
zones = pd.read_csv('D:/traffic-congestion-predictor/data/raw/taxi_zone_lookup.csv')
zones.head()
```

```python
# Merge pickup zone details
df = df.merge(
    zones,
    how='left',
    left_on='PULocationID',
    right_on='LocationID'
)

# Rename for pickup, just to avoid confusion later if any
df.rename(columns={
    'Borough': 'PUBorough',
    'Zone': 'PUZone',
    'service_zone': 'PUServiceZone'
}, inplace=True)

# Drop duplicate LocationID column, can also assign the new df without inplace as well
df.drop(columns=['LocationID'], inplace=True)
```

```python
# Merge drop-off zone details
df = df.merge(
    zones,
    how='left',
    left_on='DOLocationID',
    right_on='LocationID'
)


df.rename(columns={
    'Borough': 'DOBorough',
    'Zone': 'DOZone',
    'service_zone': 'DOServiceZone'
}, inplace=True)

df.drop(columns=['LocationID'], inplace=True)
```

```python
df_manhattan = df[
    (df['PUBorough'] == 'Manhattan') | (df['DOBorough'] == 'Manhattan')
].copy()

print("Shape after Manhattan filtering:", df_manhattan.shape)
df_manhattan.head()
```

```python
df_manhattan[['PULocationID', 'PUBorough', 'PUZone', 'DOLocationID', 'DOBorough', 'DOZone']].head(10)
```

```python
#trips that start and enf in manhattan
internal = df_manhattan[
    (df_manhattan['PUBorough'] == 'Manhattan') &
    (df_manhattan['DOBorough'] == 'Manhattan')
]

print("Manhattan → Manhattan trips:", internal.shape[0])
```

```python
# Trips that start OR end in Manhattan but not both
cross_borough = df_manhattan[
    (df_manhattan['PUBorough'] != df_manhattan['DOBorough'])
]

print("Cross-borough Manhattan trips:", cross_borough.shape[0])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# this converts pickup datetime to datetime format(already in this format) if in string format 
#df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
#df.info()

# Extract pickup hour and day of week

df['pickup_hour'] = df['tpep_pickup_datetime'].dt.hour
df['pickup_day'] = df['tpep_pickup_datetime'].dt.day_name()

# Extract dropoff hour and day of week
df['dropoff_hour'] = df['tpep_dropoff_datetime'].dt.hour
df['dropoff_day'] = df['tpep_dropoff_datetime'].dt.day_name()
print(df[['pickup_day', 'dropoff_day']].head())
```

```python
#hourly pickup and drop
pickup_hourly = df['pickup_hour'].value_counts().sort_index()
dropoff_hourly = df['dropoff_hour'].value_counts().sort_index()

# Combine them
hourly_counts = pickup_hourly.add(dropoff_hourly, fill_value=0)
```

```python
#pickup and dropoffs in days
pickup_day = df['pickup_day'].value_counts()#.sort_index()
dropoff_day = df['dropoff_day'].value_counts()#.sort_index()

# Combine
day_counts = pickup_day.add(dropoff_day, fill_value=0)
print(df[['pickup_day', 'dropoff_day']].head())
print(df['pickup_day'].dtype) 
```

```python
# Count number of pickups and dropoffs per day
#pickup_day_counts = df['pickup_day'].value_counts()
#dropoff_day_counts = df['dropoff_day'].value_counts()

# Combine them
#total_day_counts = pickup_day_counts.add(dropoff_day_counts, fill_value=0)

# Order the days
ordered_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
total_day_counts = day_counts.reindex(ordered_days)

# Converting to DataFrame for easier plotting with seaborn
day_df = total_day_counts.reset_index()
day_df.columns = ['day', 'count']

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(
    data=day_df,
    x='day',
    y='count',
    hue='day',
    palette='pastel',
    legend=False
)
plt.title("Pickup and Drop-off Frequency by Day", fontsize=14)
plt.xlabel('Day of the Week')
plt.ylabel('Pickup and Drop-off Counts')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("D:/traffic-congestion-predictor/outputs/plots/frequency_by_day.png")
plt.show()

```

```python
# Order by hour (0 to 23)
total_hour_counts = hourly_counts.sort_index()

# Convert to DataFrame
hour_df = total_hour_counts.reset_index()
hour_df.columns = ['hour', 'count']
```

```python
#Hourly bar plot with gradient colors:
import numpy as np #might be needed later

# Normalize the counts to map to colors
norm = plt.Normalize(hour_df['count'].min(), hour_df['count'].max())
colors = sns.color_palette("RdYlGn_r", len(hour_df))  # reversed: red yellow green

# Map normalized counts to color scale
mapped_colors = [colors[int(norm(c) * (len(colors)-1))] for c in hour_df['count']]

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(
    x='hour',
    y='count',
    hue='hour',
    data=hour_df,
    width= 0.8,
    palette=mapped_colors,
    legend= False
)
plt.title('Pickup and Drop-off Frequency by Hour of Day', fontsize=14)
plt.xlabel('Hour of the Day (0–23)')
plt.ylabel('Pickup and Drop-off Counts')
plt.xticks(np.arange(0, 24, 1))
plt.tight_layout()
plt.savefig('D:/traffic-congestion-predictor/outputs/plots/frequency_by_hour.png')
plt.show()
```

##Heatmap 

```python
# crosstabs for pickup and dropoff
pickup_heat = pd.crosstab(df['pickup_day'], df['pickup_hour'])
dropoff_heat = pd.crosstab(df['dropoff_day'], df['dropoff_hour'])

# Combine both
total_heat = pickup_heat.add(dropoff_heat, fill_value=0)

# Reordering the days of the week
ordered_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
total_heat = total_heat.reindex(ordered_days)

# Plot the heatmap
plt.figure(figsize=(16, 6))
sns.heatmap(total_heat, cmap='YlOrRd', linewidths=0.5, linecolor='gray')
plt.title("Pickup and Dropoff Frequency Heatmap by Day and Hour", fontsize=14)
plt.xlabel("Hour of Day")
plt.ylabel("Day of Week")
plt.tight_layout()

plt.savefig("D:/traffic-congestion-predictor/outputs/plots/heatmap_day_hour.png")
plt.show()
```
